```python
#read dataset
import pandas as pd

df = pd.read_csv(r"D:\CodeBasics Datasets\orders.csv", encoding='ISO-8859-1',na_values=['Not Available', 'unknown'])
df.head()
```

```python
#preprocessing
df['Ship Mode'].unique()
```

```python
# renaming for best practices 
#df.rename(columns={'Order Id':'order_id')
df.columns = df.columns.str.lower()  # converting to small named cols and then removing space
df.columns =  df.columns.str.replace(' ','_')
df.columns
```

```python
df.head(5)
```

```python
#derive new columns
df['discount'] = df['list_price']*df['discount_percent']*0.01
df['sale_price'] = df['list_price']-df['discount']
df['profit'] = df['sale_price']-df['cost_price']
df
```

```python
df.dtypes #check for date datatype change it to string for operations
df['order_date'] = pd.to_datetime(df['order_date'],format="%Y-%m-%d")
df.dtypes
```

```python
#dropping unneccesary columns
df.drop(columns = ['list_price','discount_percent','cost_price']) #if you want remove rows specify index =...
        
```

```python
df #so use inplace = True so remove from the dataset
```

```python
#connecting to sql
import urllib
import sqlalchemy as sal

# Build params
params = urllib.parse.quote_plus(
    "DRIVER={ODBC Driver 18 for SQL Server};"
    "SERVER=NAMEETH\\SQLEXPRESS;"
    "DATABASE=master;"
    "Trusted_Connection=yes;"
    "Encrypt=no;"
)

# Create engine with odbc_connect
engine = sal.create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# Test connection
with engine.connect() as conn:
    print("âœ… Connection successful")
```

```python
df.to_sql('df_orders', con=engine, index=False, if_exists='append') #append bcoz data type issues in sql so using append istead of replace
```
